```python

p0 = np.array([0.0, 0.0, 0.0]) 
p1 = np.array([-30.0, 0.0, 0.0]) 
p2 = np.array([0, 0, 30]) 

z = -100
pi4 = 78.78679656

po = np.array([-z, z, z])
pl = np.array([-z, z + 1, z])


list_x, list_y, list_e = list(), list(), list()
START, STOP, N, = -200, 200, 100

for x in tqdm(np.linspace(START, STOP, N)):
    for y in np.linspace(START, STOP, N):
        list_x.append(x)
        list_y.append(y)

        l_point = count_local_coords(p0, p1, p2, x, y)

        e = count_e(
            I_RGB,
            po, pl, l_point, p0, p1, p2,
            r_koef=1e10
        )
        list_e.append(e)

px.scatter(x=list_x, y=list_y, color=list_e, width=700, height=700)
```
